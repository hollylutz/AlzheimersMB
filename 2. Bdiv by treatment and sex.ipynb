```r
library(phyloseq)
library(ggplot2)
library(vegan)
library(readr)
library(data.table)
library(dplyr)
library(stats)
library(ggpubr)
library(dunn.test)
```

```r
#Read in data
alz = readRDS("/gpfs/data/gilbert-lab/Lutz/Alzheimers/RDSfiles/alz.rds") 
alz_m = subset_samples(alz, Sex=="M")
alz_f = subset_samples(alz, Sex="F")

#Rarefied data
alz_t_1k = readRDS("/gpfs/data/gilbert-lab/Lutz/Alzheimers/RDSfiles/alz_t_1k.rds")
alz_t_5k = readRDS("/gpfs/data/gilbert-lab/Lutz/Alzheimers/RDSfiles/alz_t_1k.rds")
alz_t_10k = readRDS("/gpfs/data/gilbert-lab/Lutz/Alzheimers/RDSfiles/alz_t_1k.rds")
```

#### Unweighted UniFrac

```r
#Both Sexes
alz.dist.unif = distance(alz, method = "unifrac", type = "samples")
alz.dist.unif.log = ordinate(alz, method = "PCoA", distance = "unifrac")
alz.dist.unif.evals = alz.dist.unif.log$values$Eigenvalues
#Male
alz.m.dist.unif = distance(alz_m, method = "unifrac", type = "samples")
alz.m.dist.unif.log = ordinate(alz_m, method = "PCoA", distance = "unifrac")
alz.m.dist.unif.evals = alz.m.dist.unif.log$values$Eigenvalues
#Female
alz.f.dist.unif = distance(alz_f, method = "unifrac", type = "samples")
alz.f.dist.unif.log = ordinate(alz_f, method = "PCoA", distance = "unifrac")
alz.f.dist.unif.evals = alz.f.dist.unif.log$values$Eigenvalues
```

```r
library(viridis)
p = plot_ordination(alz_f, alz.f.dist.unif.log, color = "Sex_Treatment", label="X.SampleID") + 
    ggtitle("Sex by Treatment Group") +
    geom_point(size=2, alpha=0.4) + 
    scale_colour_viridis(option="viridis", discrete=TRUE, direction=1) +
    stat_ellipse(level=0.95, geom="polygon",alpha = .2, aes(fill = Sex_Treatment), linetype=0) +
    scale_fill_viridis(option="viridis", discrete=TRUE, direction=1)+
    coord_fixed(sqrt(alz.f.dist.unif.evals[2] / alz.f.dist.unif.evals[1]))

p = p + guides(colour = guide_legend(title = "Treatment", ncol = 2, keywidth = 1, keyheight = 1))

p = p + theme_bw() + 
        theme(panel.grid.major = element_blank(),
              axis.title.y=element_text(margin=margin(0,20,0,0)),
              text=element_text(size=8, color="black",family="Arial"),
              panel.grid.minor = element_blank(),
              #axis.ticks = element_blank(),
              #legend.key.size = unit(.1, "in"),
              #legend.spacing.x = unit(.01,"in"),
              legend.position = "bottom")


p

png('sex_f_by_treatment.unif.png', width=6, height=5, units='in', res=300)
plot(p)
dev.off()
```

### Bdiv with 5k rarefied

```r
#Unweighted UniFrac
alz_t_5k.dist.unif = distance(alz_t_5k, method = "unifrac", type = "samples")
alz_t_5k.dist.unif.log = ordinate(alz_t_5k, method = "PCoA", distance = "unifrac")
alz_t_5k.dist.unif.evals = alz_t_5k.dist.unif.log$values$Eigenvalues
#Weighted UniFrac
alz_t_5k.dist.wunif = distance(alz_t_5k, method = "wunifrac", type = "samples")
alz_t_5k.dist.wunif.log = ordinate(alz_t_5k, method = "PCoA", distance = "wunifrac")
alz_t_5k.dist.wunif.evals = alz_t_5k.dist.wunif.log$values$Eigenvalues
```

#### Unweighted Unifrac PERMANOVA

```r
#Difference between treatment group within sex
alz5k.df = data.frame(sample_data(alz_t_5k))
alz5k.unif.adtest = adonis2(alz_t_5k.dist.unif ~ Treatment, strata = Sex, data=alz5k.df)
alz5k.unif.adtest
#Differences between sexes within treatment group
alz5k.df = data.frame(sample_data(alz_t_5k))
alz5k.unif.adtest = adonis2(alz_t_5k.dist.unif ~ Sex, strata = Treatment, data=alz5k.df)
alz5k.unif.adtest
```

#### Weighted Unifrac PERMANOVA

```r
#Difference between treatment group within sex
alz5k.df = data.frame(sample_data(alz_t_5k))
alz5k.wunif.adtest = adonis2(alz_t_5k.dist.wunif ~ Treatment, strata = Sex, data=alz5k.df)
alz5k.wunif.adtest
#Differences between sexes within treatment group
alz5k.df = data.frame(sample_data(alz_t_5k))
alz5k.wunif.adtest = adonis2(alz_t_5k.dist.wunif ~ Sex, strata = Treatment, data=alz5k.df)
alz5k.wunif.adtest
```
